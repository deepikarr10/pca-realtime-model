# PCA Analysis with real-time data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Step 1: Load the wine dataset
wine = load_wine()
X = wine.data           # 13 features
y = wine.target         # 3 classes (0, 1, 2)
target_names = wine.target_names
```

```python
# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_scaled.shape
```

We create a PCA object with n_components=2 to reduce to 2 dimensions

fit_transform() does two things:

- Learns the top 2 principal components (directions of highest variance)

- Projects the original data onto those components

- Output X_pca is a 2D array (178 samples Ã— 2 components)



```python
# Step 3: Apply PCA to reduce to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

Line-by-Line Explanation:

ðŸ”¹ for i, name in enumerate(target_names):

enumerate() gives us both:

i = index (0, 1, 2)

name = class name ('class_0', 'class_1', 'class_2')

So the loop runs 3 times like this:

1st iteration â†’ i=0, name='class_0'

2nd iteration â†’ i=1, name='class_1'

3rd iteration â†’ i=2, name='class_2'

ðŸ”¹ X_pca[y == i, 0]

This selects the x-values (PC1) of data points where the class label y equals i.

So:

When i=0, it selects all data points belonging to class 0 (class_0)

ðŸ”¹ X_pca[y == i, 1]

Same logic, but this gives the y-values (PC2) of those same points.

```python
# Step 4: Visualize the PCA result
plt.figure(figsize=(8, 6))
colors = ['crimson', 'limegreen', 'royalblue']
for i, name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1],
                label=name, color=colors[i])

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Wine Dataset (13D â†’ 2D)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Variance Retained:", sum(pca.explained_variance_ratio_))
```
